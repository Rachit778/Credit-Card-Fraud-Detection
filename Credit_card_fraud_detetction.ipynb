### LOGISTIC REGRESSION

```python
#TRAIN_TEST_SPLIT
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
#BUILDING LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
LOR=LogisticRegression()
LOR.fit(x_train,y_train)
```

```python
y_pred = LOR.predict(x_test)
y_pred
```

```python
#GETTING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,precision_score,recall_score,f1_score
a = confusion_matrix(y_test,y_pred)
mc_accuracy = accuracy_score(y_test,y_pred)
mc_precision = precision_score(y_test,y_pred)
mc_recall = recall_score(y_test,y_pred)
mc_f1 = f1_score(y_test,y_pred)
print("ACCURACY SCORE : ", mc_accuracy)
print(sb.heatmap(a,annot=True))
print(classification_report(y_test,y_pred))
```

### DECISION TREE CLASSIFIER

```python
#BUILDING DECISION TREE MODEL
from sklearn.tree import DecisionTreeClassifier
Treemodel = DecisionTreeClassifier(max_depth = 2)
Treemodel.fit(x_train,y_train)
```

```python
#PLOTTING TREE MODEL WITH MAX DEPTH 2
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(Treemodel,filled = True)
```

```python
#BUILDING DECISION TREE MODEL WITHOUT PARAMETER
from sklearn.tree import DecisionTreeClassifier
Treemodel = DecisionTreeClassifier()
Treemodel.fit(x_train,y_train)
```

```python
#PLOTTING TREE MODEL
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(Treemodel,filled = True)
```

```python
#PREDICTION BY THE MODEL
y_pred = Treemodel.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test,y_pred)
```

```python
#DETERMINING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
a = confusion_matrix(y_test,y_pred)
dtc_accuracy = accuracy_score(y_test,y_pred)
dtc_precision = precision_score(y_test,y_pred)
dtc_recall = recall_score(y_test,y_pred)
dtc_f1 = f1_score(y_test,y_pred)
print("ACCURACY SCORE : ",dtc_accuracy)
print(sb.heatmap(a,annot = True))
print(classification_report(y_test,y_pred))
```

### RANDOM FOREST CLASSIFIER

```python
#BUILDING RANDOM FOREST MODEL
from sklearn.ensemble import RandomForestClassifier
rfmodel = RandomForestClassifier(n_estimators=50,random_state=0)
rfmodel.fit(x_train,y_train)
```

```python
#PREDICTION BY THE MODEL
y_pred=rfmodel.predict(x_test)
y_pred
```

```python
#ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
a = confusion_matrix(y_test,y_pred)
rfc_accuracy = accuracy_score(y_test,y_pred)
rfc_precision = precision_score(y_test,y_pred)
rfc_recall = recall_score(y_test,y_pred)
rfc_f1 = f1_score(y_test,y_pred)
print("ACCURACY SCORE : ",rfc_accuracy)
print(sb.heatmap(a,annot = True))
print(classification_report(y_test,y_pred))
```

### K-NEAREST NEIGHBORS


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
#PREDICTION BY THE MODEL
y_pred=knn.predict(x_test)
y_pred
```

```python
# DETERMINING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
y_pred = knn.predict(x_test)
knn_accuracy = accuracy_score(y_test,y_pred)
knn_precision = precision_score(y_test,y_pred)
knn_recall = recall_score(y_test,y_pred)
knn_f1 = f1_score(y_test,y_pred)
print("ACCURACY SCORE : ", knn_accuracy)
print(sb.heatmap(confusion_matrix(y_test,y_pred),annot = True))
print(classification_report(y_test,y_pred))
```

### NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

#fitting-
gnb.fit(x_train, y_train)
```

```python
#predicting
y_pred=gnb.predict(x_test)
y_pred
```

```python
# DETERMINING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score
y_pred = gnb.predict(x_test)
nvb_accuracy = accuracy_score(y_test,y_pred)
nvb_precision = precision_score(y_test,y_pred,average='micro')
nvb_recall = recall_score(y_test,y_pred,average='micro')
nvb_f1 = f1_score(y_test,y_pred,average='micro')
print("ACCURACY SCORE : ", nvb_accuracy)
print(sb.heatmap(confusion_matrix(y_test,y_pred),annot = True))
print(classification_report(y_test,y_pred))
```

### SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC
svm = SVC(C=1,gamma=0.0001,kernel = 'linear')
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
y_pred
```

```python
svm_accuracy = accuracy_score(y_test,y_pred)
svm_precision = precision_score(y_test,y_pred)
svm_recall = recall_score(y_test,y_pred)
svm_f1 = f1_score(y_test,y_pred)
print("ACCURACY SCORE : ", svm_accuracy)
print(sb.heatmap(confusion_matrix(y_test,y_pred),annot = True))
print(classification_report(y_test,y_pred))
```

```python
a = pd.DataFrame({'Model' : ['Logistic Regression','Decision Tree','Random Forest','KNN','Naive Bayes','Support Vector Machine'],
                  'Accuracy' : [mc_accuracy,dtc_accuracy,rfc_accuracy,knn_accuracy,nvb_accuracy,svm_accuracy],
                  'Precision' : [mc_precision,dtc_precision,rfc_precision,knn_precision,nvb_precision,svm_precision],
                  'Recall' : [mc_recall,dtc_recall,rfc_recall,knn_recall,nvb_recall,svm_recall],
                  'F-1 Score' : [mc_f1,dtc_f1,rfc_f1,knn_f1,nvb_f1,svm_f1]})

a.style.set_table_styles([{'selector' : '',
                            'props' : [('border',
                                        '1px solid black')]}])
```
